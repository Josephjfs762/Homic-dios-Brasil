```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
taxa_homicídios=pd.read_excel("ipeadata[05-06-2025-11-23].xls")
taxa_homicídios
```

###Ajustes na base de dados

```python
# Converter de wide para long
df_long = taxa_homicídios.melt(
    id_vars=["Sigla", "Codigo", "Estado"],
    var_name="ano",
    value_name="taxa_por_100k"
)
# Remover espaços e converter para int
df_long["ano"] = df_long["ano"].astype(str).str.strip()
df_long["ano"] = pd.to_numeric(df_long["ano"], errors="coerce")
df_long
```

###Visualização Gráfica

```python
# Facetas com 5 colunas por linha (ajustável)
g = sns.FacetGrid(df_long, col="Estado", col_wrap=5, height=3.5, sharey=True)
g.map(sns.lineplot, "ano", "taxa_por_100k")
g.set_titles("{col_name}")
g.set_axis_labels("Ano", "Taxa por 100 mil")
g.fig.subplots_adjust(top=0.95)
g.fig.suptitle("Evolução da Taxa de Homicídios por Estado (1980–2023)", fontsize=16)
plt.show()
```

###Como a taxa de homicídios evoluiu no Brasil e nos estados desde 1980?


```python
inicio = df_long[df_long["ano"] == 1980]
fim = df_long[df_long["ano"] == 2023]
```

```python
# Garantir que os dataframes estejam indexados corretamente
inicio = inicio.set_index("Estado")["taxa_por_100k"]
fim = fim.set_index("Estado")["taxa_por_100k"]

# Só manter estados presentes em ambos os anos
estados_comuns = inicio.index.intersection(fim.index)
inicio = inicio.loc[estados_comuns]
fim = fim.loc[estados_comuns]

# Diferença
comparacao = fim - inicio
comparacao = comparacao.sort_values()

# Gráfico
import matplotlib.pyplot as plt

comparacao.plot(kind="barh", figsize=(10, 12), color="skyblue", title="Variação da Taxa de Homicídios (2023 - 1980)")
plt.xlabel("Diferença da Taxa por 100 mil")
plt.tight_layout()
plt.show()
```

Nossa base de dados oferece as taxas de homicídios de 1980 até 2023. Visando entender como que a Taxa de Homicídios ao longo do período em questão variou, calculamos a variação da taxa de homicídios de 1980 e de 2023 de cada estado.
Assim, observamos quais estados tiveram um aumento de suas taxas no longo prazo e quais estados tiveram suas taxas reduzidas no longo prazo.

###Quais estados têm as maiores e menores taxas em cada ano?

```python
def mostrar_taxas_por_ano(df, ano):
    if 'Estado' not in df.columns:
        print("A coluna 'Estado' não existe no DataFrame.")
        return

    df_ano = df[df['ano'] == ano]

    if df_ano.empty:
        print(f"Ano {ano} não encontrado no DataFrame.")
        return

    df_ano = df_ano.sort_values(by='taxa_por_100k', ascending=False)

    for _, row in df_ano.iterrows():
        estado = row['Estado']
        taxa = row['taxa_por_100k']
        print(f"{estado}: {taxa:.2f}")
```

```python
mostrar_taxas_por_ano(df_long, 1980)
```

```python
mostrar_taxas_por_ano(df_long, 2023)
```

###Existem padrões regionais persistentes?


```python
regioes = {
    'AC': 'Norte', 'AL': 'Nordeste', 'AP': 'Norte', 'AM': 'Norte', 'BA': 'Nordeste',
    'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste', 'GO': 'Centro-Oeste',
    'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'MG': 'Sudeste',
    'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul', 'PE': 'Nordeste', 'PI': 'Nordeste',
    'RJ': 'Sudeste', 'RN': 'Nordeste', 'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte',
    'SC': 'Sul', 'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
}

df_long['regiao'] = df_long['Sigla'].map(regioes)
```

```python
media_regional = df_long.groupby(['ano', 'regiao'])['taxa_por_100k'].mean().reset_index()
media_regional
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.lineplot(data=media_regional, x='ano', y='taxa_por_100k', hue='regiao', marker='o')
plt.title('Evolução da taxa de homicídios por 100 mil habitantes por região (1980–2023)')
plt.ylabel('Taxa por 100 mil habitantes')
plt.xlabel('Ano')
plt.legend(title='Região')
plt.grid(True)
plt.tight_layout()
plt.show()
```
